```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline


def pdf(x):
    mean = np.mean(x)
    std = np.std(x/2)
    y_out = 1/(std * np.sqrt(2 * np.pi)) * np.exp( - (x - mean)**2 / (2 * std**2))
    return y_out
  
# To generate an array of man values
x = np.arange(-5, 5, 0.1)
# To generate an array of woman values

xx= np.arange(-4.9, 5.1, 0.1)
  
# To generate a bell 
y = pdf(x)
yy=pdf(xx)

  
# Plotting the bell-shaped curve
plt.style.use('seaborn')
plt.figure(figsize = (6, 6))
plt.plot(x, y, color = 'blue',
         linestyle = 'solid',linewidth='5')
plt.plot(xx,yy,color='magenta',linestyle='solid',linewidth='5')
  
#plt.scatter(x, y, marker = 'o',
 #           s = 25, color = 'red')
  
#plt.fill_between(x_fill, y_fill, 0,
 #                alpha = 0.2, color = 'blue')
plt.show()
```
